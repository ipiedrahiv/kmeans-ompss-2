```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def create_arrays(n_samples, n_features, centers, cluster_std, center_box, random_state):
    X, y, c = make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, cluster_std=cluster_std, center_box=center_box, random_state=random_state, return_centers=True)
    return X, y, c
```

# Uniform

## 2D

### k = 7

```python
# Create a dataset with uniform, 2d, no-overlap, k = 7 clusters
X, y, c = create_arrays(10000, 2, 7, 0.1, [-5, 5], 1)

np.savetxt("uniform_small_2d_lowstddev/uniform_small_2d_lowstddev_samples.csv", X, delimiter=",")
np.savetxt("uniform_small_2d_lowstddev/uniform_small_2d_lowstddev_labels.csv", y, delimiter=",")
np.savetxt("uniform_small_2d_lowstddev/uniform_small_2d_lowstddev_centers.csv", c, delimiter=",")
```

```python
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], c=y))
sns.scatterplot(x=df["x"], y=df["y"], 
                hue=df["c"], 
                palette=sns.color_palette("tab10", n_colors=7))

plt.show()
```
